<p style="font-size:30px;">Exploratory Data Analysis</p>

<p style="font-size:20px;">Amount spent basis of Gender</p>

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x = "Gender", y = 'Amount', data = sales_gen)
```

```python
df.columns
```

<p style="font-size:20px;">Age group and Gender </p>

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = "Gender")

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

plt.show()
```

<p style="font-size:20px;">Amount spent basis of Age Group</p>

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x = "Age Group", y = 'Amount', data = sales_age)
```

<p style="font-size:20px;">Number of orders by top 10 states</p>

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x = "State", y = 'Orders', data = sales_state)
```

<p style="font-size:20px;">Amount spent by top 10 States</p>

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x = "State", y = 'Amount', data = sales_state)
```

<p style="font-size:20px;">Count of data points based on marital status</p>

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

plt.show()
```

<p style="font-size:20px;">Count of data points based on Marital status & Gender</p>

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = "Marital_Status", y = 'Amount', hue = 'Gender')
```

<p style="font-size:20px;">Count of data points based on Occupation</p>

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

plt.show()
```

<p style="font-size:20px;">Count Amount spent based on Occupation</p>

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

<p style="font-size:20px;">Count of data points based on Product_Category</p>

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

plt.show()
```

<p style="font-size:20px;">Count of data points based on Product_Category</p>

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(5)

sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')
```

<p style="font-size:20px;">Count of Orders placed based on Product_ID</p>

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```
